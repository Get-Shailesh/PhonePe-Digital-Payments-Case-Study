```python
datasets = {
    'State_Txn and Users'   : state_txn,
    'State_TxnSplit'        : state_split,
    'State_DeviceData'      : state_dev,
    'District_Txn and Users': dist_txn,
    'District Demographics' : dist_demo,
}

for name, df in datasets.items():
    print(f"\n{'='*60}\nDataset: {name}\n{'='*60}")
    print("\n-- Data Types --")
    display(df.dtypes.to_frame(name='Data Type'))
    print("\n-- Summary Statistics --")
    display(df.describe())
```

```python
print("="*60)
print("Missing Value Analysis")
print("="*60)

for name, df in datasets.items():
    missing_count = df.isnull().sum()
    missing_count = missing_count[missing_count > 0]
    print(f"\n>>> {name}")
    if missing_count.empty:
        print("  ✅ No missing values found.")
    else:
        missing_pct = (missing_count / len(df) * 100).round(2)
        summary = pd.DataFrame({'Missing Count': missing_count, 'Missing %': missing_pct})
        display(summary)
        worst_col = missing_pct.idxmax()
        print(f"  ⚠️  Highest missing %: '{worst_col}' → {missing_pct[worst_col]}%")
```

```python
# Total transactions and amount per state over all years
state_summary = state_txn.groupby('State').agg(
    Total_Transactions=('Transactions', 'sum'),
    Total_Amount_INR=('Amount (INR)', 'sum')
).reset_index()

print("Total Transactions & Amount by State:")
display(state_summary)

# Top 5 states - highest transaction volume
top5 = state_summary.nlargest(5, 'Total_Transactions')
print("\n🔝 Top 5 States — Highest Transaction Volume:")
display(top5)

# Bottom 5 states - lowest transaction volume
bottom5 = state_summary.nsmallest(5, 'Total_Transactions')
print("\n🔻 Bottom 5 States — Lowest Transaction Volume:")
display(bottom5)
```

```python
# Most frequent transaction type = the one with highest transaction count
most_common_txn = state_split.groupby(['State', 'Quarter', 'Transaction Type'])['Transactions'].sum().reset_index()

most_common_txn = most_common_txn.loc[
    most_common_txn.groupby(['State', 'Quarter'])['Transactions'].idxmax()
].reset_index(drop=True)

most_common_txn.columns = ['State', 'Quarter', 'Most Common Transaction Type', 'Transactions']

print("Most Common Transaction Type per State & Quarter:")
display(most_common_txn)
```

```python
# Step 1 — Aggregate district-level data to state level
district_agg = dist_txn.groupby('State').agg(
    District_Total_Transactions=('Transactions', 'sum'),
    District_Total_Amount=('Amount (INR)', 'sum'),
    District_Total_Users=('Registered Users', 'sum')
).reset_index()

# Step 2 — Aggregate state-level data
state_agg = state_txn.groupby('State').agg(
    State_Total_Transactions=('Transactions', 'sum'),
    State_Total_Amount=('Amount (INR)', 'sum'),
    State_Total_Users=('Registered Users', 'sum')
).reset_index()

# Step 3 — Merge both
comparison = pd.merge(district_agg, state_agg, on='State', how='outer')

# Step 4 — Calculate differences
comparison['Txn_Diff']    = comparison['State_Total_Transactions'] - comparison['District_Total_Transactions']
comparison['Amount_Diff'] = comparison['State_Total_Amount']       - comparison['District_Total_Amount']
comparison['Users_Diff']  = comparison['State_Total_Users']        - comparison['District_Total_Users']

print("Full Comparison — District vs State Level:")
display(comparison)

# Step 5 — Display discrepancies (where difference is not 0)
discrepancies = comparison[
    (comparison['Txn_Diff'] != 0) |
    (comparison['Amount_Diff'].abs() > 1) |  # small tolerance for float rounding
    (comparison['Users_Diff'] != 0)
]

print(f"\n⚠️  States with discrepancies: {len(discrepancies)}")
if discrepancies.empty:
    print("✅ No discrepancies found — district and state data are consistent!")
else:
    display(discrepancies[['State', 'Txn_Diff', 'Amount_Diff', 'Users_Diff']])
```

```python
import matplotlib.pyplot as plt

# Aggregate total population per state from district demographics
state_population = dist_demo.groupby('State')['Population'].sum().reset_index()
state_population.columns = ['State', 'Total_Population']

# Aggregate total registered users per state
state_users = state_txn.groupby('State')['Registered Users'].max().reset_index()
state_users.columns = ['State', 'Registered_Users']

# Merge both
user_population = pd.merge(state_users, state_population, on='State', how='inner')

# Calculate ratio
user_population['User_to_Population_Ratio'] = (
    user_population['Registered_Users'] / user_population['Total_Population']
).round(4)

print("Ratio of Registered Users to Population by State:")
display(user_population)

# Column chart
plt.figure(figsize=(16, 6))
plt.bar(user_population['State'], user_population['User_to_Population_Ratio'], color='steelblue')
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('User to Population Ratio')
plt.title('Ratio of Registered Users to Population by State')
plt.tight_layout()
plt.show()
```

```python
# Aggregate total transactions per district across all quarters
dist_txn_total = dist_txn.groupby(['State', 'District'])['Transactions'].sum().reset_index()
dist_txn_total.columns = ['State', 'District', 'Total_Transactions']

# Merge with district demographics on State + District
dist_merged = pd.merge(dist_txn_total, dist_demo[['State', 'District', 'Density']],
                       on=['State', 'District'], how='inner')

# Calculate correlation
correlation = dist_merged['Density'].corr(dist_merged['Total_Transactions'])
print(f"Correlation between Population Density and Transaction Volume: {correlation:.4f}")

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(dist_merged['Density'], dist_merged['Total_Transactions'] / 1e6,
            alpha=0.5, color='coral')
plt.xlabel('Population Density (per sq km)')
plt.ylabel('Total Transactions (Millions)')
plt.title('Population Density vs Transaction Volume by District')
plt.tight_layout()
plt.show()
```

```python
# Aggregate total amount and max registered users per state
state_amount = state_txn.groupby('State').agg(
    Total_Amount=('Amount (INR)', 'sum'),
    Total_Users=('Registered Users', 'max')
).reset_index()

# Calculate average transaction amount per user
state_amount['Avg_Txn_Amount_Per_User'] = (
    state_amount['Total_Amount'] / state_amount['Total_Users']
).round(2)

print("Average Transaction Amount per User by State:")
display(state_amount[['State', 'Total_Amount', 'Total_Users', 'Avg_Txn_Amount_Per_User']])

# Top 5 highest
print("\n🔝 Top 5 States — Highest Avg Transaction Amount per User:")
display(state_amount.nlargest(5, 'Avg_Txn_Amount_Per_User')[['State', 'Avg_Txn_Amount_Per_User']])

# Top 5 lowest
print("\n🔻 Top 5 States — Lowest Avg Transaction Amount per User:")
display(state_amount.nsmallest(5, 'Avg_Txn_Amount_Per_User')[['State', 'Avg_Txn_Amount_Per_User']])
```

```python
# Step 1 — Merge State_DeviceData with State_Txn and Users
device_merged = pd.merge(state_dev, 
                         state_txn[['State', 'Year', 'Quarter', 'Registered Users']],
                         on=['State', 'Year', 'Quarter'], how='left')

# Step 2 — Calculate ratio of brand users to total registered users per state
device_merged['Brand_Usage_Ratio'] = (
    device_merged['Registered Users_x'] / device_merged['Registered Users_y']
).round(4)

# Aggregate ratio per state and brand across all quarters
brand_ratio = device_merged.groupby(['State', 'Brand'])['Brand_Usage_Ratio'].mean().reset_index()
brand_ratio.columns = ['State', 'Brand', 'Avg_Usage_Ratio']
brand_ratio['Avg_Usage_Ratio'] = brand_ratio['Avg_Usage_Ratio'].round(4)

print("Device Brand Usage Ratio per State:")
display(brand_ratio)

# Step 3 — Bar chart (aggregated across all states)
overall_brand = brand_ratio.groupby('Brand')['Avg_Usage_Ratio'].mean().reset_index()
overall_brand = overall_brand.sort_values('Avg_Usage_Ratio', ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(overall_brand['Brand'], overall_brand['Avg_Usage_Ratio'], color='steelblue')
plt.xticks(rotation=45)
plt.xlabel('Device Brand')
plt.ylabel('Average Usage Ratio')
plt.title('Device Brand Usage Ratio Across All States')
plt.tight_layout()
plt.show()
```
